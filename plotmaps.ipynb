```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
seattle = pd.read_pickle('./data/seattle_routes.pklz')
```

```python
seattle.columns
```

```python
grp = seattle.groupby('routenum')

fig = plt.subplots(figsize=(15,10))

for k,g in grp:
    plt.plot(g.latitude,g.longitude,alpha=0.2,color='gray')

plt.axis('off')
plt.show()
```
